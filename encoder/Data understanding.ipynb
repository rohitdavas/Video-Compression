```python
import cv2
from glob import glob
import numpy as np
from pprint import pprint
```

# train images

```python
folder = './data'
files = glob(folder +'/train/*')

print(len(files))
pprint(files)
```

```python
def show(img):
    cv2.imshow('image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
for img in files[0]:
    img = cv2.imread(img)
    show(img)
```

```python
folder = './data/'

files = glob(folder + 'train_mv/*')

print(len(files))
pprint(files)
```
